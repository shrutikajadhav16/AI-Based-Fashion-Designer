```python
import torch
from diffusers import StableDiffusionPipeline
import gradio as gr

# Load Stable Diffusion v2.0 pipeline
model_id = "stabilityai/stable-diffusion-2"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Use GPU for inference

# Gradio function to generate images
def generate_image(prompt, guidance_scale=7.5, num_inference_steps=50):
    with torch.autocast("cuda"):
        image = pipe(prompt, guidance_scale=guidance_scale, num_inference_steps=num_inference_steps).images[0]
    return image

# Define Gradio interface
gr_interface = gr.Interface(
    fn=generate_image,
    inputs=[
        gr.Textbox(label="Prompt", placeholder="Describe the image you want"),
        gr.Slider(label="Guidance Scale", minimum=5.0, maximum=15.0, value=7.5),
        gr.Slider(label="Inference Steps", minimum=10, maximum=100, step=5, value=50),
    ],
    outputs=gr.Image(label="Generated Image"),
    title="AI Based Fasion Designer",
    description="Generate designs with Stable Diffusion v2.0 by entering a prompt and adjusting optional settings."
)

# Launch the Gradio app
gr_interface.launch()
```
